### Step 1. Import the necessary libraries.

```python
import pandas as pd
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep = "|")
users
```

### Step 4. Discover what is the mean age per occupation.

```python
round(users.groupby("occupation")["age"].agg(["mean"]).add_suffix(" of age"), 2)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

```python
male = users[users["gender"] == "M"][ "occupation"].agg(["value_counts"]) 
gender = users[["occupation", "gender"]]["occupation"].agg(["value_counts"])
ratio = male / gender

ratio.sort_values(by = "value_counts", ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages.

```python
users.groupby("occupation")["age"].agg(["min", "max"]).add_suffix("imum age")
```

### Step 7. For each combination of occupation and gender, calculate the mean age.

```python
round(users.groupby(["occupation", "gender"])["age"].agg(["mean"]).add_suffix(" age"), 2)
```

### Step 8.  For each occupation present the percentage of women and men.

```python
pd.crosstab(users["occupation"], users["gender"]).apply(lambda r:round(r / r.sum() * 100, 2), axis = 1)
```
